# **Deskriptive Datenanalyse**

```
import pandas as pd
import numpy as np
import seaborn as sns
```

**Daten laden**

```
data = pd.read_csv("https://tinyurl.com/wfn4clj", sep=";")

data.head()
```

**Punktschätzer für unsere Daten (Mean, Median & Mode)**
---



**Mean - Mittelwert (oder auch arimetrische Mitte) 
(Summe aller Daten geteilt durch die Anzahl der Datensätze)** Für metrische, intervall oder ratioskalierte Daten sehr gut geeignet. *Bei ordinalen Werten weniger geeigent, ausser es wird mit Integerwerten gearbeitet*

```
data.age.mean()
```

**Median (Der Wert in der Mitte bei im Beispiel nach Alter aufsteigend sortierten Datensätzen)** - Bei ordinalen Daten besonders geeignet (Hier liegen 50% der Werte darunter und 50% der Werte darüber. Hier spielen Ausreißer eine untergeordnete Rolle.) Geeignet um festzustellen, ob es starke Ausreißer gibt.

```
data.age.median()
```

**Mode (der Wert der Datensätze, welcher am Häufigsten vorkommt)** Besonders für Kategorische Daten (Häufigste Kategorie z.B. Geschlecht, Kundenkategorie, ...)

```
data.age.mode()
```



---



**Streuungsschätzer für unsere Daten ()**
---



**Minimaler Wert**

```
data.age.min()
```

**Maximaler Wert**

```
data.age.max()
```

**Qantiler Wert** 
- Getrimmte Range (ein bestimmter Wert) z.B. 0.9 = 90% Hier schneiden wir, je nach Kurve, die "Enden/Ausreißer bzw. Long Tails" in diesem Beispiel die obere und untere 5% weg. Die Abbildungsgenauigkeit wird exakter.

```
data.age.quantile(q=0.9)
```

**Standard deviation - Standardabweichung**
Nur für metrische Merkmale gut geeignet. Und in Kombination mit dem arimetrischen Mittel.

```
data.age.std()
```

In unserem Beispiel können wir sagen der Mittelwert ist 41,17 Jahre +/- der Standardabweichung von 10,5 Jahre, die sie rauf oder runter streuen. Die meisten Kunden sind zwischen 30 - 50 Jahre alt.
In dieser Range liegen 67% aller Werte (Normalverteilung). 
Wenn wir **zwei Standardaweichungen** verwenden dann liegt die Range bei 95%.
Bei **drei Standardabweichungen** liegt die Range schon bei 97% aller Werte.
